#1) Write a function to reverse the nodes of a given linked list.

```
#include<stdio.h>
#include<stdlib.h>
#include<conio.h>
struct Node{
    int Data;
    struct Node *Next;
}*S,*L,*C,*P;
void insertNode();
void deleteNode();
void Display();
void reverseDisplay();
int main()
{
    int x;
    S=L=C=P=NULL;
    while(1)
    {
        printf("\n Single Linked List ");
        printf("\n 1. Insert ");
        printf("\n 2. Delete ");
        printf("\n 3. Display ");
        printf("\n 4. Reverse Display ");
        printf("\n 5. Exit ");
        printf("\n Enter Your Choice: ");
        scanf("%d",&x);
        switch(x)
        {
            case 1:insertNode();
            break;
            case 2:deleteNode();
            break;
            case 3:Display();
            break;
            case 4:reverseDisplay();
            break;
            case 5:printf("\n Program is terminating...");
            return 0;
            break;
        }
    }
    return 0;
}
void insertNode()
{
    struct Node *nn;
    nn=(struct Node *)malloc(sizeof(struct Node));
    printf("\n Enter the value : ");
    scanf("%d",&nn->Data);
    nn->Next=NULL;
    if(S==NULL)
    {
        S=L=nn;
    }
    else
    {
        L->Next=nn;
        L=nn;
    }
    printf("\n Node is inserted at end");
    getch();
}
void deleteNode(){
    if(S==NULL){
		printf("\n Single Linked List is empty");
	}
	else{
		if(S->Next == NULL){
			S = L = NULL;
		}
		else{
			C = S;
			S = S->Next;
			C->Next = NULL;
			C = NULL;
		}
		printf("\n Node is deleted from start of Linked List");
	}
}
void reverseDisplay()
{


    if(S!=NULL)
    {
        P=S;
        C=S->Next;
        S=S->Next;
        P->Next=NULL;
        while(S!=NULL)
        {
            S=S->Next;
            C->Next=P;
            P=C;
            C=S;
        }
        S=P;
    }
    Display();
}
void Display()
{
    if(S==NULL)
    {
        printf("\n Single Linked List is empty");
        getch();
    }
    else
    {
        printf("\n Nodes in Linked List is : \n");
        for(C=S;C!=NULL;C=C->Next)
        {
            printf("%d ",C->Data);
        }
        getch();
    }
}
```

#2) Write a  C program to remove vowels from given linked list.

Expected Output : b-> c-> f-> g

```
#include<stdio.h>
#include<stdlib.h>
#include<conio.h>
#include<stdbool.h>
struct Node{
    char Data;
    struct Node *Next;
}*S,*L,*C,*P;
void insertNode();
void deleteNode();
void Display();
bool isVowel(char x)
{
    return (x == 'a' || x == 'e' || x == 'i'
            || x == 'o' || x == 'u' || x == 'A'
            || x == 'E' || x == 'I' || x == 'O'
            || x == 'U');
}
int main()
{
    int x;
    S=L=C=P=NULL;
    while(1)
    {
        printf("\n Single Linked List ");
        printf("\n 1. Insert ");
        printf("\n 2. Delete ");
        printf("\n 3. Display ");
        printf("\n 4. Exit ");
        printf("\n Enter Your Choice: ");
        scanf("%d",&x);
        switch(x)
        {
            case 1:insertNode();
            break;
            case 2:deleteNode();
            break;
            case 3:Display();
            break;
            case 4:printf("\n Program is terminating...");
            return 0;
            break;
        }
    }
    return 0;
}
void insertNode()
{
    struct Node *nn;
    nn=(struct Node *)malloc(sizeof(struct Node));
    printf("\n Enter the value : ");
    scanf("%s",&nn->Data);
    nn->Next=NULL;
    if(S==NULL)
    {
        S=L=nn;
    }
    else
    {
        L->Next=nn;
        L=nn;
    }
    printf("\n Node is inserted at end");
    getch();
}
void deleteNode(){
	struct Node* ptr=S;
	while(ptr!=NULL)
    {
        if(isVowel(ptr->Data))
            ptr=ptr->Next;
        else
            break;
    }
    P=ptr;
    S=ptr;
    while(ptr!=NULL)
    {
        if(isVowel(ptr->Data))
        {
            ptr=ptr->Next;
            while(ptr!=NULL)
            {
                if(isVowel(ptr->Data))
                {
                    ptr=ptr->Next;
                }
                else
                    break;
            }
            if(ptr==NULL)
            {
                P->Next=NULL;
                break;
            }
            else
            {
                P->Next=ptr;
            }
        }
        P=P->Next;
        ptr=ptr->Next;
    }
}

void Display()
{
    if(S==NULL)
    {
        printf("\n Single Linked List is empty");
        getch();
    }
    else
    {
        printf("\n Nodes in Linked List is : \n");
        for(C=S;C!=NULL;C=C->Next)
        {
            printf("%c ",C->Data);
        }
        getch();
    }	
}
```
