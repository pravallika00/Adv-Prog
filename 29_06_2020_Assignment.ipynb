```
#include<stdio.h>
#include<stdlib.h>
struct Node
{
    int n;
    struct Node *Next;
}*S;

void listCreation(int nodes);
void insertNode(int n,int pos);
void display(int a);

int main()
{
    int nodes,val,a,x;
    S=NULL;
    printf("Input the number of nodes : ");
    scanf("%d",&nodes);
    listCreation(nodes);
    a=1;
    display(a);
    printf("Input the position to insert a new node : ");
    scanf("%d",&x);
    printf("Input data for the position %d : ",x);
    scanf("%d",&val);
    insertNode(val,x);
    a=2;
    display(a);
    return 0;
}

void listCreation(int nodes)
{
    int i,n;
    struct Node *P,*nn;
    if(nodes>=1)
    {
        S=(struct Node *)malloc(sizeof(struct Node));
        printf("Input data for node 1 : ");
        scanf("%d",&n);
        S->n=n;
        S->Next=NULL;
        P=S;
        for(i=2;i<=nodes;i++)
        {
            nn=(struct Node *)malloc(sizeof(struct Node));
            printf("Input data for node %d : ",i);
            scanf("%d",&n);
            nn->n=n;
            nn->Next=NULL;
            P->Next=nn;
            P=nn;
        }
        P->Next=S;
    }
}

void insertNode(int n,int pos)
{
    struct Node *nn,*C;
    int i;
    if(pos==1)
    {
        nn=(struct Node *)malloc(sizeof(struct Node));
        nn->n=n;
        nn->Next=S;
        C=S;
        while(C->Next!=S)
        {
            C=C->Next;
        }
        C->Next=nn;
        S=nn;
    }
    else
    {
        nn=(struct Node *)malloc(sizeof(struct Node));
        nn->n=n;
        C=S;
        for(i=2;i<=pos-1;i++)
        {
            C=C->Next;
        }
        nn->Next=C->Next;
        C->Next=nn;
    }
}

void display(int l)
{
    struct Node *t;
    int nodes=1;
    if(S==NULL)
    {
        printf("List is empty");
    }
    else
    {
        t=S;
        if(l==1)
        {
            printf("\n Data entered in the list are : \n");
        }
        else
        {
            printf("After insertion the new list is : \n");
        }
        do
        {
            printf("Data %d = %d\n",nodes,t->n);
            t=t->Next;
            nodes++;
        }while(t!=S);
    }
}
```
